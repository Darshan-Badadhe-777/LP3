```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay,recall_score,precision_score
```

```python
X=df.drop(df['Outcome'])
Y=df['Outcome']
print(X.shape)
print(Y.shape)
#Idhar dono k size alag h matlab hame koi min len wala choose krna padega 
min_len=min(len(X),len(Y))
X=X[:min_len]
Y=Y[:min_len]
```

```python
kneighbors=KNeighborsClassifier(n_neighbors=5)
```

```python
kneighbors
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
kneighbors.fit(X_train,Y_train)
```

```python
k_ypredict=kneighbors.predict(X_test)
```

```python
k_ypredict
```

```python
accuracy=accuracy_score(Y_test,k_ypredict)
recall=recall_score(Y_test,k_ypredict)
precision=precision_score(Y_test,k_ypredict)
ConfusionMatrixDisplay.from_predictions(Y_test,k_ypredict)
report=classification_report(Y_test,k_ypredict)
```

```python
print("Overall : ")
print("Accuracy_score" , accuracy)
print("Recall :" ,recall)
print("Precision : ",precision)
print("Report :")
print(report)
```
